```python
#Transformasi data menghapus kolom yang tidak digunakan
df_Cbanking.drop(['emp_var_rate', 'cons_price_idx', 'cons_conf_idx', 'euribor3m', 'nr_employed', 'y'], axis = 1)
```

```python
#Filter data yang digunakan 
dfCbanking = df_Cbanking[['job', 'education', 'duration']]
dfCbanking
```

```python
#Cek Kategori data kolom job
dfCbanking['job'].value_counts()
```

```python
#Cek kategori kolom education
dfCbanking['education'].value_counts()
```

```python
#Labelling Data
list_status = []

for i, j in dfCbanking.iterrows():
    if ((j['job'] == 'management') and (j['education'] == 'university.degree') and (j['duration'] <150)) :
        segment = 'rich'
    elif ((j['job'] == 'technician') and (j['education'] == 'university.degree') and (j['duration'] > 150 and j['duration'] < 200)):
        segment = 'middle'
    elif ((j['job'] == 'blue-collar') and (j['education'] == 'basic.4y' or j['education'] == 'basic.6y' or j['education'] == 'basic.9y') and (j['duration'] >200)) :
        segment = 'poor'
    else :
        segment = 'unknown'
    list_status.append(segment)
    
dfCbanking['Status'] = list_status
dfCbanking['Status'].value_counts()
```
